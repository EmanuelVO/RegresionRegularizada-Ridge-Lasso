```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel("FuelConsumptionCo2.xlsx")
df.head()
```

```python
#Remover caracteristicas categoricas
df.drop(['MODELYEAR', 'MAKE', 'MODEL', 'VEHICLECLASS', 'TRANSMISSION', 'FUELTYPE'], axis=1, inplace=True)
df
```

```python
#Remover renglones con valores faltantes
df.dropna(inplace=True)
df
```

```python
X = df.drop('CO2EMISSIONS', axis=1)
y = df['CO2EMISSIONS']
```

```python
# Division en grupos de entrenamiento(training) y de prueba(testing)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#Normalizacion
from sklearn.preprocessing import normalize
X_train_norm = normalize(X_train)
X_test_norm = normalize(X_test)
```

# Regresion lineal multiple

```python
# Construccion de un modelo de regresion lineal multiple
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(X_train_norm, y_train)
```

```python
# Coeficientes de regresion
print('Intercepto: ', linreg.intercept_)
print('Coeficientes: ', linreg.coef_)
```

```python
valores_x = X.iloc[0]

intercepto = linreg.intercept_
coeficientes = linreg.coef_
```

```python
# Pronostico de la primera observación de la base de prueba
y_pronosticada = intercepto + sum(coeficientes * valores_x)
y_pronosticada
```

```python
# Predicciones
y_pred = linreg.predict(X_test_norm)
y_pred
```

```python
# Calculo de indicadores de la bondad de ajuste
from sklearn.metrics import r2_score
from sklearn import metrics
```

```python
# Impresion de indicadores de la bondad de ajuste
print('Valor de R2: ', r2_score(y_test, y_pred))
print('Error absoluto medio: ', metrics.mean_absolute_error(y_test, y_pred))
print('Error cuadratico medio: ', metrics.mean_squared_error(y_test, y_pred))
print('Raiz del error cuadratico medio: ', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

# Ridge

```python
alpha_range = 10.**np.arange(-2, 3)
alpha_range
```

```python
from sklearn.linear_model import RidgeCV

ridgeregcv = RidgeCV(alphas=alpha_range, scoring='neg_mean_squared_error')
ridgeregcv.fit(X_train_norm, y_train)
ridgeregcv.alpha_
```

```python
# Coeficientes de regresion
print('Intercepto: ', ridgeregcv.intercept_)
print('Coeficientes: ', ridgeregcv.coef_)
```

```python
intercepto = ridgeregcv.intercept_
coeficientes = ridgeregcv.coef_
```

```python
# Pronostico de la primera observación de la base de prueba
y_pronosticada = intercepto + sum(coeficientes * valores_x)
y_pronosticada
```

```python
# Predicciones
y_pred = ridgeregcv.predict(X_test_norm)
y_pred
```

```python
# Impresion de indicadores de la bondad de ajuste
print('Valor de R2: ', r2_score(y_test, y_pred))
print('Error absoluto medio: ', metrics.mean_absolute_error(y_test, y_pred))
print('Error cuadratico medio: ', metrics.mean_squared_error(y_test, y_pred))
print('Raiz del error cuadratico medio: ', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

# Lasso
